<h2>9-1 '한국복지패널 데이터' 분석 준비하기</h2>
<br/><br/>한국복지패널 데이터 : 한국보건사회연구원에서 우리나라 가구의 경제활동을 연구해 복지 정책에 반영할 목적으로 발간하는 조사 자료. 전국에서 7,000여 가구를 선정해 2006년부터 매년 추적 조사했으며 쳔여 개 변수로 구성되어 있음.  

```python
#pyreadstat 패키지 설치 (vs code에서 pip 사용이 안되서 아나콘다 프롬프트에서 해당 명령어 입력하여 설치함)
#pip install pyreadstat
import pyreadstat

# 데이터를 분석하는데 필요한 패키지 로드하기
import pandas as pd
import numpy as np
import seaborn as sns

# 데이터 불러오기
raw_welfare = pd.read_spss('D:/WE_IT/doit/bookdata/Koweps_hpwc14_2019_beta2.sav')

# 복사본 만들기 
welfare = raw_welfare.copy()
```

```python
# 데이터 검토하기
welfare   # 앞부분, 뒷부분 출력
```

```python
welfare.shape  # 행, 열 개수 출력
```

```python
welfare.info()  # 변수 속성 출력
```

```python
welfare.describe()  # 요약 통계량
```

변수의 수가 많고 변수명이 코드로 되어 있어 전체 구조를 파악하기 어려움. <br/>-> 변수명을 쉬운 단어로 바꾼 다음 분석에 사용할 변수를 하나씩 살펴봐야 함. <br/><br/>
변수명을 바꿀 때는 ❗코드북❗을 활용

```python
# 변수명 바꾸기
welfare = welfare.rename(
    columns = {'h14_g3'      : 'sex',            # 성별
               'h14_g4'      : 'birth',          # 태어난 연도
               'h14_g10'     : 'marriage_type',  # 혼인 상대
               'h14_g11'     : 'religion',       # 종교
               'p1402_8aq1'  : 'income',         # 월급
               'h14_eco9'    : 'code job',       # 직업 코드
               'h14_reg7'    : 'code region'})   # 지역 코드
```

<h3>[데이터 분석 절차]</h3>
<br/>데이터 분석은 두 단계로 진행됨.
<br/> **1단계 변수 검토 및 전처리 -> 2단계 변수 간 관계 분석**
<br/><br/> 이때 1단계에서는 분석에 활용할 변수 각각에 대해 전처리를 진행해야 함.
<br/> 2단계에서는 데이터를 요약한 표와 데이터의 특징을 쉽게 이해할 수 있는 그래프를 만든 후 분석 결과를 해석해야 함.
