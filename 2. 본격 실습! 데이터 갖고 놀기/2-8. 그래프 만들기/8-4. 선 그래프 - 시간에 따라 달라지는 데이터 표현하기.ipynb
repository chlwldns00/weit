```python
import pandas as pd
import seaborn as sns

economics = pd.read_csv("/Users/heohyeonjun/Desktop/We_IT_5th/weit/Data/economics.csv") # 데이터 불러오기
```

```python
sns.lineplot(data = economics, x = 'date', y = 'unemploy')
```

```python
# 날짜 시간 타입 변수 만들기
economics['date2'] = pd.to_datetime(economics['date'])

# 변수 타입 확인
economics.info()
```

```python
# 연도 변수 추가
economics['year'] = economics['date2'].dt.year
economics.head()
```

```python
# x축에 연도 표시
sns.lineplot(data = economics, x = 'year', y = 'unemploy')
```

```python
# 신뢰구간 제거
sns.lineplot(data = economics, x = 'date2', y = 'unemploy', ci=None)
```
