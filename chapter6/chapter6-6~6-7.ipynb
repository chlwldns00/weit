<h2>6-6 집단별로 요약하기</h2>

```python
import pandas as pd
exam = pd.read_csv('D:/WE_IT/doit/bookdata/exam.csv')
exam
```

전체 요약 통계량 구하기

```python
# 수학 점수 평균을 구해 mean_math에 할당하는 기능
exam.agg(mean_math = ('math', 'mean'))
```

집단별 요약 통계량 구하기

```python
# 반별 수학 점수 평균 구하기 (nclass별로 분리 후 math 평균 구함)
exam.groupby('nclass') \
		.agg(mean_math = ('math', 'mean'))
```

```python
# 변수를 인덱스로 바꾸지 않기
exam.groupby('nclass', as_index = False) \
		.agg(mean_math = ('math', 'mean'))
```

여러 요약 통계량 한 번에 구하기

```python
exam.groupby('nclass') \
		.agg(mean_math   = ('math', 'mean'),      # 수학 점수 평균
		     sum_math    = ('math', 'sum'),       # 수학 점수 합계
		     median_math = ('math', 'median'),    # 수학 점수 중앙값
		     n           =('nclass', 'count'))    # 빈도(학생 수)
```

모든 변수의 요약 통계량 한 번에 구하기

```python
exam.groupby('nclass').mean()
```

mpg 데이터로 실습

```python
mpg = pd.read_csv('D:/WE_IT/doit/bookdata/mpg.csv')
mpg
```

```python
# 제조 회사별로 집단을 나눈 다음 다시 구동 방식별로 나눠 도시 연비 평균을 구함
mpg.groupby(['manufacturer', 'drv']) \
   .agg(mean_cty = ('cty', 'mean'))
```

출력 결과 drv의 행의 개수가 제조 회사별로 다른데, 그 이유는 회사마다 생산한 drv의 종류가 다르기 때문!

```python
# audi의 drv별 빈도 (audi추출, drv별 분리, 빈도 구하기)
mpg.query('manufacturer == "audi"')\
	 .groupby(['drv']) \
	 .agg(n = ('drv', 'count'))
```

```python
# chevrolet의 drv별 빈도
mpg.query('manufacturer == "chevrolet"')\
	 .groupby(['drv']) \
	 .agg(n = ('drv', 'count'))
```

agg()와 count()가 아닌 df.value_counts()를 이용하여 집단별 빈도 구하는 법

```python
# 빈도를 구할 변수를 추출한 다음 value_counts()를 적용하면 됨.
mpg['drv'].value_counts()
```

```python
# 시리즈 자료 구조인 출력 결과에 query()를 적용하기 위해 데이터 프레임으로 변환해야 함.
mpg['drv'].value_counts() \
				  . to_frame('n') \
				  .rename_axis('drv') \
				  .query('n > 100')		
				 	
```

pandas 함수 조합하기

```python
mpg.query('category == "suv"') \
	 .assign(total = (mpg['hwy'] + mpg['cty']) / 2) \
	 .groupby('manufacturer') \
	 .agg(mean_tot = ('total', 'mean')) \
	 .sort_values('mean_tot', ascending = False) \
	 .head()
```

<h2>6-7 데이터 합치기</h2>

가로로 합치기

```python
# 실습 위한 데이터 프레임 2개 만들기
# 중간고사 데이터 만들기
test1 = pd.DataFrame({'id'        : [1, 2, 3, 4, 5],
                      'midterm'   : [60, 80, 70, 90, 85]})
                      
# 기말고사 데이터 만들기
test2 = pd.DataFrame({'id'        : [1, 2, 3, 4, 5],
                      'final'     : [70, 83, 65, 95, 80]})

print(test1)
print(test2)
```

```python
# id 기준으로 두 시험 데이터를 합쳐 하나의 데이터 프레임 total에 할당
total = pd.merge(test1, test2, how = 'left', on = 'id')
total
```

세로로 합치기

```python
# 실습 위한 데이터 프레임 2개 만들기
# 학생 1~5번 시험 데이터 만들기
group_a = pd.DataFrame({'id'       : [1, 2, 3, 4, 5],
                        'test'     : [60, 80, 70, 90, 85]})
                        
# 학생 6~10번 시험 데이터 만들기
group_b = pd.DataFrame({'id'       : [6, 7, 8, 9, 10],
                        'test'     : [70, 83, 65, 95, 80]})

print(group_a)
print(group_b)
```

```python
# 데이터 합쳐서 group_all에 할당
group_all = pd.concat([group_a, group_b])
group_all
```

```python
# 인덱스 새로 부여
group_all = pd.concat([group_a, group_b], ignore_index = True)
group_all
```
